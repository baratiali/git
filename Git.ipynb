# Git

`git inint` 

`ls -ltrha .git`: see the directories

`git status`

`git log`: shows the logs

`git add "page*`: add all the files with name page (page1 , page2, ...)


`git diff HEAD`: [explains the diffrence](https://www.atlassian.com/de/git/tutorials/saving-changes/git-diff)

`git diff --staged`: show what changed in files the in stage are

`git reset page3.html`

`git checkout --page3.html`: revert it to the last changed

`git brnach branchname`: creating a new branch

`git branche`: shows list of the branches

`git checkout branchname`: change the branch to the other one

`git merg subbranchname`: merges the subbranch to master branch

`git rm filename`: delet it from filesystem and git

`git branch -d branch name`: delete a branch
